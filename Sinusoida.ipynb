```python
import matplotlib.pyplot as plt
from matplotlib.backends.backend_tkagg import FigureCanvasTkAgg
import numpy as np
import tkinter as tk
import customtkinter
customtkinter.set_appearance_mode("dark")
customtkinter.set_default_color_theme("dark-blue")

def update_plot(*args):
    amplitude = amplitude_slider.get()
    frequency = frequency_slider.get()
    x = np.linspace(-10, 10, 1000)
    y = amplitude * np.sin(frequency * x)
    ax.clear()
    ax.plot(x, y)

    ax.set_ylim(-10, 10)
    ax.set_xlim(-10, 10)
    ax.grid(True, which='both', linestyle=':', linewidth=0.1)

    canvas.draw()

root = customtkinter.CTk()
root.grid_rowconfigure(0, weight=1)
root.grid_columnconfigure(0, weight=1)

fig, ax = plt.subplots()
canvas = FigureCanvasTkAgg(fig, master=root)
canvas.get_tk_widget().grid(row=0, column=0, sticky="nsew", padx=10, pady=10)

amplitude_slider = customtkinter.CTkSlider(master=root, from_=1, to=10, number_of_steps=9, command=update_plot)
amplitude_slider.set(2)
amplitude_slider.grid(row=1, column=0, sticky="ew", padx=10)

frequency_slider = customtkinter.CTkSlider(master=root, from_=1, to=50, number_of_steps=49, command=update_plot)
frequency_slider.set(10)
frequency_slider.grid(row=2, column=0, sticky="ew", padx=10)

update_plot()  # Narysuj poczÄ…tkowy wykres

root.mainloop()
```
